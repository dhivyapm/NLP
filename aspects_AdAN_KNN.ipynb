```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_squared_error
from math import sqrt
df = pd.read_csv('aspects_AdAN_sent.csv',sep=',')
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
cat_variables = df[['aspects', 'emotions']]
cat_dummies = pd.get_dummies(cat_variables, drop_first=True)
cat_dummies.head()
```

```python
df = df.drop(['aspects', 'emotions'], axis=1)
df = pd.concat([cat_dummies,df], axis=1)
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =0)
```

```python
R_squared = []
RMSE = []
model = KNeighborsRegressor(n_neighbors = 20, p=1, metric = 'euclidean')
model.fit(X_train, y_train)
prediction = model.predict(X_test)
r2 = r2_score(y_test, prediction)
rmse = sqrt(mean_squared_error(y_test, prediction))
R_squared.append(r2)
RMSE.append(rmse)
regr = ['KNN']
basic_result = pd.DataFrame({'R squared':R_squared,'RMSE':RMSE},index = regr)
basic_result
```
